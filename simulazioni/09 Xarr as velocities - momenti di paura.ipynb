## Compute the trajs using scn['Xarr'] as velocities

```python
# Create the dataset using scene
df = dflist[0]
scn_list = seq2scn(df)
```

```python
scn = scn_list[0]
v0 = 30
```

```python
N, tstamps = scn['N. vehicles'], scn['Tarr']
formatter = '{0:.02f}'

print("--"*50)
print(f"Scene n. {scn.name}, time interval: [{formatter.format(tstamps[0])},{formatter.format(tstamps[-1])}]")

x_exact, t_exact, v_exact = [[0]]*(N), [scn['Tarr'][0]], []

for i in range(0,len(tstamps)-1):

    print("--"*50)
    print(f"Time interval n.{i}: [{formatter.format(scn['Tarr'][i])}, {formatter.format(scn['Tarr'][i+1])}]\n")

    ### TRUE
    t0, tend, deltat = tstamps[i], tstamps[i+1], 0.05
    tspan_ann_true = time_discretization(t0,tend,deltat)

    tmp = [traj[-1] for traj in x_exact]
    x0 = np.array(tmp).flatten().tolist()       

    # abuso del modello TD_ANN_model
    vel = scn['Xarr'][:,i]
    sol_ann_true = odeint(TD_ANN_model, x0, tspan_ann_true, args=(vel,), tfirst = True).T

    ## STEP 3: Update x_ann_list, t_ann_list
    x_ann = sol_ann_true.tolist()
    t_ann = tspan_ann_true[1:] # avoid the first recording

    # add sol to the correct veh
    for j in range(0,N):
        tmp = x_ann[j][1:] # avoid the first recording
        x_exact[j] = np.concatenate([x_exact[j],tmp])
    t_exact = np.concatenate([t_exact,t_ann]).tolist()
    
print("=="*50)
```

```python
tstamps = scn['Tarr']
N = scn['N. vehicles']

# fig 
width, height = 7, 5
fig, ax = plt.subplots(figsize=(width,height))

for veh in range(0,N):
    ax.plot(t_exact,x_exact[veh])

for ts in tstamps:
    ax.axvline(x = ts, color = 'red', linestyle = '--', linewidth=0.75)

# Annotate the name of each variable at the last value
# plot_annotation(ax, scn, trajs, scale)

ax.set_xlabel("t")
ax.set_ylabel("X(t)")
ax.set_title(r"$X_{arr}\ as\ velocities$", fontsize=15)
# plt.legend(labels_plot)
plt.grid()
plt.show()
```
