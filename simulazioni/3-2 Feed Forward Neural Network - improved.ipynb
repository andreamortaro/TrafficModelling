# Import libraries

```python
# Python libraries
import os
import pandas as pd # for data manipulation
import numpy as np # for data manipulation
import tensorflow as tf
import matplotlib.pyplot as plt

# To enable LaTeX and select a font
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.sans-serif": "Helvetica",
})

# Using ipynb import functions defined in other nb
from ipynb.fs.defs.myfun_load_dataset import *
from ipynb.fs.defs.myfun_models import *
from ipynb.fs.defs.myfun_plot import *
```

# Data Acquisition

```python
# Load the dataset
par_dir = os.path.dirname(os.getcwd()) # parent dir
dir_name = par_dir + "/NN-interaction"
merged_df, dflist = load_dataset(dir_name, 'post') # we have 10 pandas dataframe, one for each couple(camera,sequence)
```

```python
# For sake of simplicity conseder only a scene
df = dflist[0] # sequence
scn = df.T[0]  # scene
```

```python
# Useful params
N = scn['N. vehicles']
tstamps = scn['Tarr']
```

## Create the model NN

```python
DOE =[[1,2,1]] #Design of experiment
```

```python
model = create_model(DOE[0])
# model.summary()
```

## Evaluate the NN in each time interval

```python
# parameters to set
epochs = 100          # An epoch is an iteration over the entire x and y data provided
batch_size = 50       # Number of samples per gradient update.
v0 = 30
```

### Case tout = "end"

```python
tout="end"
t_ann_end, x_ann_end, vel_ann_end = odesolver_ann_scene(model, scn, batch_size, epochs, v0, tout, verbose='auto')
```

#### Plot

```python
# Params for the plot
trajs = [scn['Xarr'], np.array(x_ann_end)]
labels_plot = ["true","nn sim"]
title = f"Trajs simulated by NN driven LWR model\ntout = {tout}"
```

```python
# fig 
width, height = 7, 5
fig, ax = plt.subplots(figsize=(width,height))

for veh in range(0,N):
    ax.plot(tstamps, scn['Xarr'][veh])
    ax.plot(t_ann_end, x_ann_end[veh], '--')

for ts in tstamps:
    ax.axvline(x = ts, color = 'red', linestyle = '--', linewidth=0.75)
    
# Annotate the name of each variable at the last value
plot_annotation(ax, scn, trajs, scale=1.004)

ax.set_xlabel("t")
ax.set_ylabel("X(t)")
ax.set_title(title, fontsize=15)
# Limits for the plot
plot_limits(ax, scn, trajs, xbal=0.01, ybal=0.05)
plt.legend(labels_plot)
plt.grid()
plt.show()
```

### Case tout = "all"

```python
tout="all"
t_ann_all, x_ann_all, vel_ann_all = odesolver_ann_scene(model, scn, batch_size, epochs, v0, tout, verbose='Off')
```

#### Plot

```python
# Params for the plot
trajs = [scn['Xarr'], np.array(x_ann_all)]
labels_plot = ["true","nn sim"]
title = f"Trajs simulated by NN driven LWR model\ntout = {tout}"
```

```python
# fig 
width, height = 7, 5
fig, ax = plt.subplots(figsize=(width,height))

for veh in range(0,N):
    ax.plot(tstamps, scn['Xarr'][veh])
    ax.plot(t_ann_all, x_ann_all[veh], '--')

for ts in tstamps:
    ax.axvline(x = ts, color = 'red', linestyle = '--', linewidth=0.75)
    
# Annotate the name of each variable at the last value
plot_annotation(ax, scn, trajs, scale=1.004)


ax.set_xlabel("t")
ax.set_ylabel("X(t)")
ax.set_title(title, fontsize=15)
# Limits for the plot
plot_limits(ax, scn, trajs, xbal=0.01, ybal=0.05)
plt.legend(labels_plot)
plt.grid()
plt.show()
```

### Comparing tout: all and end

```python
# Initialize the figure
width = 7
height = 5
nfig = 2
fig = plt.figure(figsize = (width*nfig,height))

# 1. tout = all
ax1 = fig.add_subplot(1, nfig, 1)
for veh in range(0,N):
    ax1.plot(tstamps, scn['Xarr'][veh])
    ax1.plot(t_ann_end, x_ann_end[veh], '--')
for ts in tstamps:
    ax1.axvline(x = ts, color = 'red', linestyle = '--', linewidth=0.25)
plot_annotation(ax1, scn, trajs, scale=1.001)

ax1.set_xlabel("t")
ax1.set_ylabel("X(t)")
ax1.set_title(f"tout=end", fontsize=15)
ax1.grid(color='grey', linestyle='-', linewidth=0.5);
ax1.legend(labels_plot);

# 2. tout = end
ax2 = fig.add_subplot(1, nfig, 2);
for veh in range(0,N):
    ax2.plot(tstamps, scn['Xarr'][veh])
    ax2.plot(t_ann_all, x_ann_all[veh], '--')
for ts in tstamps:
    ax2.axvline(x = ts, color = 'red', linestyle = '--', linewidth=0.25)
plot_annotation(ax2, scn, trajs, scale=1.001)

ax2.set_xlabel("t")
ax2.set_ylabel("X(t)")
ax2.set_title(f"tout=all", fontsize=15)
ax2.grid(color='grey', linestyle='-', linewidth=0.5);
ax2.legend(labels_plot);


fig.suptitle('Comparing "tout" option',fontsize=25, y=0.95);
fig.subplots_adjust(top=0.8)
plt.show();
```
