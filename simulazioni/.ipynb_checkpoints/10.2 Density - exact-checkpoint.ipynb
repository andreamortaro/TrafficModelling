## Riemann problem

```python
# Flux

## Traffic Jam
if flux_flag == 0:
    
    f = lambda rho: rho*(1-rho)  
    fprime = lambda rho: 1-2*rho
    fprime_inv = lambda rho: .5*(1-rho)

## Burger    
if flux_flag == 1:
    
    f = lambda rho: rho*rho
    fprime = lambda rho: 2*rho
    fprime_inv = lambda rho: .5*rho
```

```python
# Solution for the Traffic Jam case
def riemmann_traffic_exact(rho_l, rho_r):
    
    if rho_r > rho_l: # Shock wave
        shock_speed = (f(rho_l)-f(rho_r))/(rho_l-rho_r)
        def reval(xi):
            if xi <= shock_speed:
                return rho_l
            else:
                return rho_r
            
    else: # Rarefaction wave
        c_l = fprime(rho_l)
        c_r = fprime(rho_r)

        def reval(xi):
            if xi <= c_l:
                return rho_l
            elif xi >= c_r:
                return rho_r
            else:
#                 return (1.-xi)/2.
                return fprime_inv(xi)
    return reval
```

## Characteristics

```python
## Characteristics

def characteristic(xspan, tspan, tol=0.1):
    
    result = []
    shock_speed = (f(rho_l)-f(rho_r))/(rho_l-rho_r)

    c_l = fprime(rho_l)
    c_r = fprime(rho_r)

    for x0 in xspan:
        
        char = [x0]
        nocross_cond = True
        car_side = int(x0 < 0)*["left"] + int(x0 > 0)*["right"]
        it = 1

        while (it < len(tspan) and nocross_cond):
            
            t = tspan[it]
            shock = shock_speed*t
            
            if char[-1] <= shock:
                tmp = x0 + c_l*t
            else:
                tmp = x0 + c_r*t
            
            # update step
            actual_side = int(tmp < shock)*["left"] + int(tmp > shock)*["right"]
            if actual_side != car_side: # already cross
                nocross_cond = False
            else: # not crossed yet
                char.append(tmp)
                nocross_cond = (abs(tmp-shock) > tol) # too near to shock
                it += 1

        # Case: intersection btw char and shock
        while (it < len(tspan)):
            
            t = tspan[it]
            shock = shock_speed*t
            tmp = shock_speed*t

            #update & storage
            char.append(tmp)
            it += 1
            
        result.append(char)
        
    return result
```

```python
def cars_trajectories(xspan, tspan, rho_l, rho_r):
    
    reval = riemmann_traffic_exact(rho_l,rho_r)
    
    shock_speed = (f(rho_l)-f(rho_r))/(rho_l-rho_r)
    u_left = 1 - rho_l
    u_right = 1 - rho_r
     
    x_trajs = []
    
    for x0 in xspan:
    
        car_traj = [x0]
        t_prec = tspan[0]
        initial_side = int(car_traj[-1] < 0)*["left"] + int(car_traj[-1] > 0)*["right"]
        current_side = initial_side

        if initial_side == ["left"]:
            u = u_left
        else:
            u = u_right

        it = 1
            
        while (it < len(tspan) and current_side == initial_side):
            
            t = tspan[it]
            tmp = car_traj[-1] + u*(t-t_prec)
            shock = shock_speed*t
            
            # update
            current_side = int(tmp < shock)*["left"] + int(tmp > shock)*["right"]

            if current_side == initial_side:
                car_traj.append(tmp)
                t_prec = t
                current_side = int(tmp < shock)*["left"] + int(tmp > shock)*["right"]
                it +=1
            else:
                if current_side == ["right"]:
                    u = u_right
                elif current_side == ["left"]:
                    u = u_left
                    
        while (it < len(tspan)):
            
            t = tspan[it]
            tmp = car_traj[-1] + u*(t-t_prec)
            
            # update & storage
            car_traj.append(tmp)
            t_prec = t
            it +=1
            
        x_trajs.append(car_traj)

    return x_trajs
```

```python
def rho(xspan, t, rho_l, rho_r):
    
    "Density of cars distributed as xspan at time=t"
    
    reval = riemmann_traffic_exact(rho_l,rho_r)
    
    result = []
    
    if t == 0:
        for x0 in xspan:
            tmp = int(x0<0)*rho_l + int(x0>=0)*rho_r
            result.append(tmp)
    else:
        for x0 in xspan:
            xi = x0/t
            tmp = reval(xi)
            result.append(tmp)
                
    return result
```

```python
ttrange = 0, T/2, T

# Cars trajectories
cars_trajs = cars_trajectories(xspan, tspan, rho_l, rho_r)
cars_trajs_transposed = list(map(list, zip(*cars_trajs)))
```
