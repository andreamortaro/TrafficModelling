## Plot

```python
# Create the model NN
DOE =[[1,2,1], [1,4,1], [1,10,1]] #Design of experiment

# params
NUM_ITER = int(dir_nn2.split("_")[-1].split("it")[0]) #same for all the models
```

### Create the out dir

```python
# Create the directory..
if flag_save:
    
    ext = ".svg"
    
    df_seen_str = dir_nn2.split("_")[2]
    
    # Create directory where to save the image
    now = datetime.now() 
    d = now.strftime(f"%Y-%m-%d_%H-%M-%S_{NUM_ITER}it-PRE-nn-all")
    
    path = 'out/' + d
    os.mkdir(path)
```

```python
ms2kmh = 3.6

for d in DOE:

    # Initialize the fig
    nrow, ncol = 1, 1
    width, height = 7, 5 # single subplot
    fig, ax = plt.subplots(nrow,ncol,figsize=(width*ncol,height))
    
    # Chose info for one DOE
    info_dataset = info_alldataset.loc[info_alldataset['DOE'] == str(d)]   

    #
    info_grouped = info_dataset.groupby(['N. file', 'type']).mean(numeric_only=True)
    info_grouped2 = info_dataset.groupby(['N. file']).mean(numeric_only=True)
    info_avg = info_dataset.groupby(['type']).mean(numeric_only=True)

    # All the data
    ## For each scene, plot v0 mean
    for _, scn in info_dataset.iterrows():
        ax.scatter(scn['N. file'], scn['v0']*ms2kmh, alpha = .2, color = 'darkblue');

    ## For a dataset, plot v0 mean (averaging v0 mean in all the scenes)
    for nf, p in info_grouped2.iterrows():
        ax.plot(nf, p['v0']*ms2kmh, color = 'r', marker="x")

    # Plot v0 mean over all dfs
    alldfs = [a for a, _ in info_dataset.groupby('N. file')]
    v0_mean = np.mean([ infof['v0']*ms2kmh for _, infof in info_grouped.iterrows()]) 
    ax.plot(alldfs, [v0_mean]*len(alldfs), color = 'r')  

    # labeling     
    ax.set_xlabel(r"$data set$",fontsize=14)
    ax.set_ylabel(r"$v_{0}\ [km/h]$",fontsize=14)
    ax.set_title(fr"$Velocities\ v_{0}\ of\ the\ leading\ car\ with\ NN\ structure\ {str(d)}$",fontsize=14)

    # Setting the values for all axes.
    custom_xlim = (0.75,10.25)
    custom_xticks = range(1,11)
    plt.setp(ax, xlim=custom_xlim, xticks=custom_xticks)

    ax.grid(color='grey', linestyle='-', linewidth=0.5);

    plt.show()
    
    # Save figure
    if flag_save:

        title = f"/v0_NN-{str(d[1]).zfill(2)}"     
        fig.savefig(path+title+ext, bbox_inches='tight')
```

```python
# Initialize the fig
nrow, ncol = 1, 3
width, height = 7, 5 # single subplot
fig, axes = plt.subplots(nrow,ncol,figsize=(width*ncol*.75,height*.75))

ms2kmh = 3.6

for step, d in enumerate(DOE):
        
    # Chose info for one DOE
    info_dataset = info_alldataset.loc[info_alldataset['DOE'] == str(d)]   

    #
    info_grouped = info_dataset.groupby(['N. file', 'type']).mean(numeric_only=True)
    info_grouped2 = info_dataset.groupby(['N. file']).mean(numeric_only=True)
    info_avg = info_dataset.groupby(['type']).mean(numeric_only=True)

    # All the data
    ## For each scene, plot v0 mean
    for _, scn in info_dataset.iterrows():
        axes[step].scatter(scn['N. file'], scn['v0']*ms2kmh, alpha = .2, color = 'darkblue');

    ## For a dataset, plot v0 mean (averaging v0 mean in all the scenes)
    for nf, p in info_grouped2.iterrows():
        axes[step].plot(nf, p['v0']*ms2kmh, color = 'r', marker="x")

    # Plot v0 mean over all dfs
    alldfs = [a for a, _ in info_dataset.groupby('N. file')]
    v0_mean = np.mean([ infof['v0']*ms2kmh for _, infof in info_grouped.iterrows()]) 
    axes[step].plot(alldfs, [v0_mean]*len(alldfs), color = 'r')  

    # labeling
    axes[step].set_title(fr"$NN\ structure\ {str(d)}$",fontsize=18)
    axes[step].grid(color='grey', linestyle='-', linewidth=0.5);

fig.supxlabel(r"$data set$",fontsize=16, y=0.001)
fig.supylabel(r"$v_{0}\ [km/h]$",fontsize=16, x = 0.075)
plt.suptitle(fr"$Velocities\ v_{0}\ of\ the\ leading\ car$", fontsize=22, y=1.075)
    
# Setting the values for all axes.
custom_xlim = (0.75,10.25)
custom_xticks = range(1,11)
plt.setp(axes, xlim=custom_xlim, xticks=custom_xticks)    

plt.show()
    
# Save figure
if flag_save:

    title = f"/v0_NN-together"     
    fig.savefig(path+title+ext, bbox_inches='tight')
```

### Plot comparing

```python
# Initialize the fig
nrow, ncol = 1, 1
width, height = 7, 5 # single subplot
fig, ax = plt.subplots(nrow,ncol,figsize=(width*ncol,height))

ms2kmh = 3.6

palette = ['green', 'royalblue', 'coral']


for step, d in enumerate(DOE):   
    
    # Choose data for a fixed doe
    info_dataset = info_alldataset.loc[info_alldataset['DOE'] == str(d)]
    
    color = palette[step]

    # For a dataset, plot v0 mean (averaging v0 mean in all the scenes)
    info_grouped = info_dataset.groupby(['N. file']).mean(numeric_only=True)
    for nf, p in info_grouped.iterrows():
        ax.scatter(nf, p['v0']*ms2kmh, facecolors='none', edgecolors=color)

    # Plot v0 mean over all dfs
    alldfs = [a for a, _ in info_grouped.iterrows()]
    v0_mean_alldfs = info_grouped['v0'].mean()*ms2kmh
    ax.plot(alldfs, [v0_mean_alldfs]*len(alldfs), color = color, label=fr"$NN {str(d[1])}$")
    

    
ax.set_xlabel(r"$data set$",fontsize=14)
ax.set_ylabel(r"$v_{0}\ [km/h]$",fontsize=14)
ax.set_title(fr"$Velocities\ v_{0}\ of\ the\ leading\ car$",fontsize=14)
ax.grid(color='grey', linestyle='-', linewidth=0.5);
ax.legend()

# Setting the values for all axes.
custom_xlim = (0.75,10.25)
custom_xticks = range(1,11)
plt.setp(ax, xlim=custom_xlim, xticks=custom_xticks)

plt.show()

# Save figure
if flag_save:

    title = f"/v0_NN-ALL-comparing"
    fig.savefig(path+title+ext, bbox_inches='tight')
```
