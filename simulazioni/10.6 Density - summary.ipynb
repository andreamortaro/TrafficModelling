# Preliminary

```python
## Params
Nx = 100
Nt = 320
T = 0.5
```

```python
# Space - linearly distributed
dx = 1/int(Nx/3)
a = np.linspace(-1,0,int(Nx/3))
b = np.linspace(0+dx/2, 1, Nx-int(Nx/3))
xspan = np.concatenate([a, b])
cons_dis = np.diff(xspan, axis=0)

# Time
tspan = np.linspace(0,T,Nt)
```

## Exact density

```python
## Params
rho_l = 0.4
rho_r = 0.8
rho_out = 0
```

```python
## Flux
#Traffic Jam
f = lambda rho: rho*(1-rho)  
fprime = lambda rho: 1-2*rho
fprime_inv = lambda rho: .5*(1-rho)

# params
toll = 1e-3
flux = [f, fprime, fprime_inv]
speed_shock = (f(rho_r) - f(rho_l))/(rho_r - rho_l)
```

```python
## plot params
N = Nx # how many chars, cars trajs to compute
xlims = [-1,1] # computational x limits
drho = 0.2
xlim1, xlim2 = -1.25, 1.75 # plot x limits

# # t to plot
# idx = 0, int(len(tspan)*.5), -1
```

```python
## Compute the velocities
y_pred = model_nn(cons_dis, training=False)

## Approximate the velocity of the leading car ?
v0 = y_pred.numpy().mean()
vel_nn = y_pred.numpy().flatten().tolist() + [v0]

# check
if all(i >= v0 for i in y_pred):
    print("There is at least a vehicle faster than the leading car") 
```

## Lin/Log

```python
# cons_dis = np.diff(xspan, axis=0)
# L = np.mean(cons_dis)
L_lin = 0.003
L_log = 0.01175
# v0 = 30
v0 = y_pred.numpy().mean()
```

## Solutions and discretized density

```python
ln = 1/Nx

# Lin
sol_lin = odeint(TD_LWR_model, xspan, tspan, args=(Nx, v0, L_lin, 'Lin'), tfirst = True).T
rho_lin = ln/np.diff(sol_lin, axis=0)

# Log
sol_log = odeint(TD_LWR_model, xspan, tspan, args=(Nx, v0, L_log, 'Log'), tfirst = True).T
rho_log = ln/np.diff(sol_log, axis=0)

# NN10
sol_ann = odeint(TD_ANN_model, xspan, tspan, args=(vel_nn,), tfirst=True).T
rho_ann = ln/np.diff(sol_ann, axis=0)
```

### Plot with different models

```python
sols = [sol_lin, sol_log, sol_ann]
rhos = [rho_lin, rho_log, rho_ann]

titles = [fr"$Lin\ case (L,v_0)={np.round(L_lin,3), np.round(v0,3)}$",
          fr"$Log\ case (L,v_0)={np.round(L_log,3), np.round(v0,3)}$",
          fr"$NN10\ case v_0={np.round(v0,3)}$"]
```

```python
# Initialize the fig
nrow, ncol = 1, 3
width, height = 7, 5 # single subplot
fig, axes = plt.subplots(nrow,ncol,figsize=(width*ncol*.75,height*.75))

for sol, ax, title in zip(sols, axes.ravel(), titles):
    # Plot the true and simulated trajectories for each vehicle
    for veh in range(0, Nx):
        ax.plot(tspan, sol[veh])
    
    ax.grid(color='grey', linestyle='-', linewidth=0.1)
    ax.set_title(title, fontsize=18, y = 1.075)

plt.suptitle(fr"$FTL\ system\ with\ f_\theta$", fontsize=20,y=1.1);
fig.supxlabel("$t$", x= 0.5, y= -0.015)
fig.supylabel("$X(t)$", x =0.085)
plt.show()
```

## Plot mischione selvaggio

```python
# Initialize the fig
nrow, ncol = 1, 3
width, height = 7, 5 # single subplot
fig, axs = plt.subplots(nrow,ncol,figsize=(width*ncol*.75,height*.75))

plt.suptitle(fr"$Discretized\ density\ comparison$", fontsize=18, y=1.05); # Figure titles

for col, (ax, i) in enumerate(zip(axs, idx)):

    t = tspan[i]

    # exact density
    idx_t, (cp_t, d_t), _, _ = rho_2s(xlims, N, tspan, rho_l, rho_r, rho_out, flux, np.round(t,8), drho)
    ax.plot(cp_t, d_t, c='g', ls = '-', lw = .5, marker='*', label="Exact density")
    if i != idx_t:
        print("warning: time stamps for exact and lin/log density don't match!")

    # Lin
    x = sol_lin[:,i][1:]
    y = rho_lin[:,i]
    ax.plot(x,y,c='red', ls='-', lw = 0.5, ms=5, marker='*', label=fr"$\rho_n$ with lin model")
#             ax.plot(x,x*0,c='blue', ls='', ms=5, marker='o', label="Particles")

    # Log
    x = sol_log[:,i][1:]
    y = rho_log[:,i]
    ax.plot(x, y, c='orange', ls='-', lw = 0.5, ms=5, marker='*', label=fr"$\rho_n$ with lin model")
#             ax.plot(x,x*0,c='blue', ls='', ms=5, marker='o', label="Particles")

    # NN
    x = sol_ann[:,i][1:]
    y = rho_ann[:,i]
    
    # Cutting off values over 1
    x_fixed, y_fixed = list(x.copy()), list(y.copy())
    
    j, rhomax = 0, 1
    
    while j<len(y_fixed):
        if rhomax-y_fixed[j] < 0:
            y_fixed.pop(j), x_fixed.pop(j)
            j = j-1
        j+=1
        
    ax.plot(x_fixed, y_fixed, c='blue', ls='-', lw = 0.5, ms=5, marker='x', label=fr"$\rho_n$ with nn model")
#             ax.plot(x,x*0,c='blue', ls='', ms=5, marker='o', label="Particles")

#     ax.set_xlim([-1.5,1.5])
    ax.set_ylim([0, 1])
    ax.set_title(fr"$t={t.round(3)}$",fontsize=16)

    ax.grid(color='grey', linestyle='-', linewidth=0.5)
    ax.legend()

fig.supxlabel("$x$", fontsize=18, y=-0.025)
fig.supylabel(fr"$\rho$", fontsize=18, x=.075)
    
plt.show()
```
